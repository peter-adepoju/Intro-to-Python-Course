## Section 1: Aliasing — Two Names, One Object

```python
a = [1, 2, 3]
b = a           # b is an ALIAS -- same object, NOT a copy

print("id(a):", id(a))
print("id(b):", id(b))
print("a is b:", a is b)    # identity test
print("a == b:", a == b)    # equality test
```

```python
# PREDICT before running -- this surprises almost everyone
a = [1, 2, 3]
b = a
b.append(4)     # mutate through b

print("b:", b)  # Prediction: ____
print("a:", a)  # Prediction: ____   <- did a change?
```

## Section 2: `is` vs `==`

```python
a = [1, 2, 3]
b = a              # alias
c = [1, 2, 3]      # independent list with same values

print("a == b:", a == b)    # True (same values)
print("a is b:", a is b)    # True (same object)

print("a == c:", a == c)    # True (same values)
print("a is c:", a is c)    # False (different objects)

# Now mutate through b and see what happens to c
b.append(4)
print("\nafter b.append(4):")
print("a:", a)   # changed?
print("c:", c)   # changed?
```

## Section 3: Cloning

```python
original = [1, 2, 3, 4, 5]

copy1 = original[:]      # full slice
copy2 = list(original)   # list() constructor
copy3 = original.copy()  # .copy() method

# All three are independent
copy1.append(99)
print("original:", original)   # unchanged
print("copy1:   ", copy1)

# Confirm independence
print("original is copy2:", original is copy2)   # False
```

## Section 4: Aliasing Through Function Calls

```python
# A function that mutates its argument
def add_zero(lst):
    lst.append(0)   # mutates the caller's list!

my_list = [1, 2, 3]
add_zero(my_list)
print(my_list)   # [1, 2, 3, 0] -- the original was modified!
```

```python
# A function that works on a clone -- original is protected
def doubled(lst):
    """Returns a NEW list with every element doubled."""
    result = []
    for x in lst:
        result.append(x * 2)
    return result

nums = [1, 2, 3]
new = doubled(nums)
print("original:", nums)   # unchanged
print("doubled: ", new)
```

## Day 24 Wrap-Up

**Key things to remember:**
1. `b = a` creates an ALIAS, not a copy -- same object in memory
2. Mutating through any alias affects all aliases
3. `is` tests identity (same object); `==` tests equality (same values)
4. Clone with `a[:]`, `list(a)`, or `a.copy()` to get an independent copy
5. Function parameters are aliases -- mutations inside functions affect the caller's list
