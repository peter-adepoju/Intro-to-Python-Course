## Section 2: Indexing, Slicing, and Item Assignment

```python
fruits = ["apple", "banana", "cherry", "date"]

print(fruits[0])      # first
print(fruits[-1])     # last
print(fruits[1:3])    # slice -- returns a NEW list
print(len(fruits))

# PREDICT before running:
scores = [95, 82, 78, 91]
scores[1] = 88        # item assignment -- NEW! tuples can't do this
print(scores)          # Prediction: ____
```

## Section 3: List Operators and Built-Ins

```python
a = [1, 2, 3]
b = [4, 5, 6]

print(a + b)        # concatenation -- new list
print(a)            # a unchanged
print(a * 3)        # repetition -- new list
print(2 in a)       # membership
print(7 not in a)   # non-membership

nums = [3, 1, 4, 1, 5, 9, 2, 6]
print(len(nums), min(nums), max(nums), sum(nums))
```

## Section 4: Iterating Over Lists

```python
fruits = ["apple", "banana", "cherry"]

# Direct iteration
for fruit in fruits:
    print(fruit)

print()

# Index-based iteration
for i in range(len(fruits)):
    print(f"{i}: {fruits[i]}")

print()

# enumerate: both index and value
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")
```

## Section 5: The List Accumulator Pattern

```python
# Build a list of squares
def squares_up_to(n):
    """
    Assumes: n is a positive integer
    Returns: list of perfect squares from 1**2 to n**2
    """
    result = []              # start empty
    for i in range(1, n+1):
        result.append(i*i)   # grow one element at a time
    return result

print(squares_up_to(6))   # [1, 4, 9, 16, 25, 36]
```

## Day 22 Wrap-Up

**Key things to remember:**
1. Lists are ordered, MUTABLE sequences: `[1, 2, 3]`
2. Item assignment (`lst[i] = val`) -- only lists (not strings or tuples) support this
3. `+` and slicing always create NEW lists; the original is unchanged
4. `len()`, `min()`, `max()`, `sum()` all work on lists
5. List accumulator: `result = []; ... result.append(x); return result`
