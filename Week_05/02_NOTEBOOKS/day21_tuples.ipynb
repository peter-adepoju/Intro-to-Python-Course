## Recap: Week 4 in 60 Seconds

```python
# Multiple return values have been tuples all along — now you know why
def min_and_max(a, b, c):
    """Assumes a, b, c are numbers. Returns (min, max) as a tuple."""
    smallest = min(a, min(b, c))   # using built-in min for clarity
    largest  = max(a, max(b, c))
    return smallest, largest         # tuple packing!

low, high = min_and_max(7, 2, 9)   # tuple unpacking!
print(low, high)
```

## Section 1: Creating Tuples

```python
# Various ways to create tuples
empty  = ()
single = (42,)         # trailing comma is REQUIRED for a 1-element tuple
pair   = (1, 2)
triple = ("a", "b", "c")
packed = 10, 20, 30    # parentheses optional -- tuple packing

print(empty, type(empty))
print(single, type(single))
print(packed, type(packed))
```

```python
# The single-element gotcha -- PREDICT before running
not_a_tuple = (42)
is_a_tuple  = (42,)

print(type(not_a_tuple))   # Prediction: ____
print(type(is_a_tuple))    # Prediction: ____
```

## Section 2: Indexing, Slicing, and Immutability

```python
colors = ("red", "green", "blue", "yellow")

print(colors[0])     # first element
print(colors[-1])    # last element
print(colors[1:3])   # slice -- returns a tuple
print(len(colors))   # length
```

```python
# Immutability: tuples cannot be modified
point = (3, 7)
try:
    point[0] = 99
except TypeError as e:
    print(f"TypeError: {e}")

# Iteration works fine
for color in colors:
    print(color)
```

## Section 3: Tuple Unpacking

```python
# Basic unpacking
point = (3, 7)
x, y = point
print(f"x={x}, y={y}")

# Swap two variables cleanly
a, b = 5, 10
a, b = b, a
print(f"after swap: a={a}, b={b}")

# Unpack a string too
first, second, third = "abc"
print(first, second, third)
```

```python
# Mismatched unpacking raises ValueError
try:
    a, b = (1, 2, 3)
except ValueError as e:
    print(f"ValueError: {e}")
```

## Section 4: Tuples as Return Values and Record Types

```python
def circle_stats(radius):
    """
    Assumes: radius is a positive number
    Returns: (area, circumference) as a tuple
    """
    pi = 3.14159
    area = pi * radius ** 2
    circumference = 2 * pi * radius
    return area, circumference

a, c = circle_stats(5)
print(f"area={a:.2f}, circumference={c:.2f}")

# Tuple as a record: (name, age, score)
student = ("Alice", 20, 92)
name, age, score = student
print(f"{name} (age {age}): {score}")
```

**Key things to remember:**
1. Tuples are ordered, IMMUTABLE sequences: `(1, 2, 3)`
2. Single-element tuples require a trailing comma: `(42,)` not `(42)`
