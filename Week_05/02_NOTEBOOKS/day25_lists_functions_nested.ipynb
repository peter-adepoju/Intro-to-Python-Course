## Section 1: Mutate In Place vs. Return New List

```python
# Style 1: mutate in place (side-effect style)
def scale(lst, factor):
    """Modifies lst in place -- multiplies every element by factor."""
    for i in range(len(lst)):
        lst[i] = lst[i] * factor

prices = [5, 10, 20, 50]
scale(prices, 2)
print(prices)   # [10, 20, 40, 100]
```

```python
# Style 2: return a new list (pure function style)
def scaled(lst, factor):
    """Returns a NEW list with every element multiplied by factor."""
    result = []
    for x in lst:
        result.append(x * factor)
    return result

prices = [5, 10, 20, 50]
new_prices = scaled(prices, 2)
print(prices)       # [5, 10, 20, 50]  -- unchanged
print(new_prices)   # [10, 20, 40, 100]
```

## Section 2: Common List-Processing Patterns

```python
# FILTER pattern: keep elements satisfying a condition
def keep_above(lst, threshold):
    result = []
    for x in lst:
        if x > threshold:
            result.append(x)
    return result

# MAP pattern: transform every element
def square_all(lst):
    result = []
    for x in lst:
        result.append(x * x)
    return result

# REDUCE pattern: combine all elements into one value
def product(lst):
    total = 1
    for x in lst:
        total *= x
    return total

nums = [5, 12, 3, 8, 1, 10]
print(keep_above(nums, 6))   # [12, 8, 10]
print(square_all([1,2,3,4])) # [1, 4, 9, 16]
print(product([2, 3, 4]))    # 24
```

## Section 3: Nested Lists

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(matrix[0])      # first row: [1, 2, 3]
print(matrix[1][2])   # row 1, col 2: 6

# PREDICT each value before running:
print(matrix[2][0])   # Prediction: ____
print(matrix[0][-1])  # Prediction: ____
```

```python
# Nested loops to process every element
def matrix_sum(matrix):
    total = 0
    for row in matrix:
        for element in row:
            total += element
    return total

def flatten(matrix):
    result = []
    for row in matrix:
        for element in row:
            result.append(element)
    return result

m = [[1, 2], [3, 4], [5, 6]]
print(matrix_sum(m))   # 21
print(flatten(m))       # [1, 2, 3, 4, 5, 6]
```

## Section 4: Recursion on Lists — Week 4 Transfers Directly

```python
# Strings:  base case = "",  recurse on s[1:]
# Lists:    base case = [],  recurse on lst[1:]  -- same pattern!

def recursive_sum(lst):
    if lst == []:
        return 0
    return lst[0] + recursive_sum(lst[1:])

def recursive_max(lst):
    if len(lst) == 1:
        return lst[0]
    rest_max = recursive_max(lst[1:])
    return lst[0] if lst[0] > rest_max else rest_max

print(recursive_sum([1, 2, 3, 4, 5]))              # 15
print(recursive_max([3, 1, 4, 1, 5, 9, 2, 6]))    # 9
```
