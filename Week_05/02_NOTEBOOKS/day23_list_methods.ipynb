## Section 1: Adding Elements

```python
# append -- adds to end, returns None
fruits = ["apple", "banana"]
fruits.append("cherry")
print(fruits)

# append a list adds it as ONE element
nums = [1, 2, 3]
nums.append([4, 5])
print(nums)   # [1, 2, 3, [4, 5]] -- not [1, 2, 3, 4, 5]!

# insert -- adds at a specific index
fruits = ["apple", "cherry"]
fruits.insert(1, "banana")
print(fruits)
```

## Section 2: Removing Elements

```python
# remove -- removes FIRST occurrence by value
fruits = ["apple", "banana", "cherry", "banana"]
fruits.remove("banana")
print(fruits)   # only the first 'banana' is removed

# pop -- removes by index AND returns the value
fruits = ["apple", "banana", "cherry"]
last = fruits.pop()     # removes last
print(last, fruits)

first = fruits.pop(0)   # removes at index 0
print(first, fruits)
```

## Section 3: The Critical Distinction -- sort() vs sorted()

```python
# THE MOST COMMON BEGINNER TRAP:
words = ["banana", "apple", "cherry"]
words = words.sort()     # ← BUG: .sort() returns None!
print(words)              # None -- your list is gone!

print()

# CORRECT -- two options:
# Option 1: sort in place (don't rebind)
words = ["banana", "apple", "cherry"]
words.sort()
print(words)   # ['apple', 'banana', 'cherry']

# Option 2: get a new sorted list
words = ["banana", "apple", "cherry"]
sorted_words = sorted(words)
print(sorted_words)   # ['apple', 'banana', 'cherry']
print(words)           # ['banana', 'apple', 'cherry'] -- unchanged!
```

## Section 4: The Mutating-vs-Non-Mutating Table

```python
lst = [3, 1, 4, 1, 5]

# These MUTATE (return None):
lst.append(9)           # adds 9 to end
lst.insert(0, 0)        # adds 0 at front
lst.remove(1)           # removes first 1
lst.sort()              # sorts in place
print("After mutations:", lst)

# This RETURNS a value (and also mutates):
removed = lst.pop()     # removes & returns last
print("Popped:", removed, "lst now:", lst)

# These do NOT mutate:
original = [3, 1, 4]
new1 = sorted(original)    # new sorted list
new2 = original + [5]      # new concatenated list
new3 = original[:]         # new copy
print("original unchanged:", original)
```

## Day 23 Wrap-Up

**Key things to remember:**
1. `.sort()`, `.append()`, `.insert()`, `.remove()`, `.reverse()` all MUTATE and return `None`
2. `.pop()` is the exception: mutates AND returns the removed element
3. `sorted()` returns a NEW list; `.sort()` mutates in place
4. NEVER write `lst = lst.sort()` -- you get `None`
5. `sorted(lst)[-n:]` is a clean way to get the top n elements
